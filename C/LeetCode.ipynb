```cpp
#include <algorithm> // 必须包含
#include <vector>
#include <iostream>

int main() {
    std::vector<int> v(5);
    // 将向量中所有元素填充为 10
    std::fill(v.begin(), v.end(), 10); // v = {10, 10, 10, 10, 10}

    int arr[5];
    // 将数组前 3 个元素填充为 5
    std::fill(arr, arr + 3, 5); // arr = {5, 5, 5, ?, ?}
}
```
